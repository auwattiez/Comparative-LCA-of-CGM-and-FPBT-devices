# Create custom activities

```python
"""
The aim here is to create activities that don't exist in Ecoinvent. They are custom activities.
Note: at the end of the day, they are still built from different activities that exist in the ecoinvent database.
"""

for sheet_name, sheet_df in custom_sheet_dataframes.items(): # iterate in the custom sheets, note : they all begin with "process..." in the excel
    print(f"Processing sheet: {sheet_name}") # to keep track of which sheet is been processed
    accumulated_exchanges = {}
    
    for index, row in sheet_df.iterrows():         # iterate inside a given custom sheet
        ei_activity_name = row["EI activity name"] # get the name of the ei activity at a given row 
        if pd.isna(ei_activity_name):              # Break when 'EI activity name' is NaN, -> when we're at the end of the i-th custom sheet
            print(f"Skipping row {index} in sheet '{sheet_name}' as 'EI activity name' is NaN.")
            break

        location = row["loc"]
        parameter_expression = row["parameter"]
        new_activity_name = row["LCA algebraic name"]

        ei_activity = agb.findTechAct(ei_activity_name, loc=location) # search in ei database
        if ei_activity is None:
            print(f"Skipping row {index} in sheet '{sheet_name}' due to unresolved activity issues.") # if the activity is not found, should not happen !
            continue

        try: # We also need to associate exchanges inside custom activities with parameter values
            parameter_value = eval(parameter_expression, {}, parameter_registry) # evaluate the parameter expression
            if ei_activity in accumulated_exchanges:
                accumulated_exchanges[ei_activity] += parameter_value  # accumulate value if activity already exists
            else:
                accumulated_exchanges[ei_activity] = parameter_value  # add new activity to exchanges
        except Exception as e:
            print(f"Error in parameter expression for activity in row {index}: {e}")
            continue

    try:
        agb.newActivity(USER_DB, sheet_name, "unit", exchanges=accumulated_exchanges) # Create the custom process with accumulated exchanges and associated parameters
        print(f"Exchanges for {sheet_name}: {accumulated_exchanges} \n") # for debugging or for a better understanding
    
    except Exception as e:
        print(f"Error creating custom activity '{sheet_name}': {e} \n") # Should not happen !

print("Finished processing all sheets.")
```

# Modified activities

```python
for index, row in foregrounds_All[f"foreground_CGM_{USER_DB}"].iterrows():
    
    ei_activity_name = row["EI activity name"]  # Read values from the Excel file
    modified_activity_name = row["modified"]
    location = row["loc"]
    
    if pd.isna(ei_activity_name):  # Stop processing when 'EI activity name' is NaN
        print(f"Stopping processing at index {index} as 'EI activity name' is NaN.")
        break  # Exit the loop
    
    if pd.notna(modified_activity_name):   
        try:
            original_activity = agb.findTechAct(ei_activity_name, loc=location)
            if original_activity is None:
                print(f"Skipping copy and update for {modified_activity_name} due to unresolved activity issues.")
                continue
            
            new_activity = agb.copyActivity(USER_DB, original_activity, modified_activity_name)  # Create a copy of the activity
            print(f"Copied activity: {modified_activity_name}")
            
            sheet_data = modified_sheet_dataframes[modified_activity_name]
            exchanges_dict = {}

            # TBM = To Be Modified inside the copied activity
            for idx, sheet_row in sheet_data.iterrows():
                ei_activity_name_TBM = sheet_row["EI activity name"]
                parameter_expression_TBM = sheet_row["parameter"]
        
                try:
                    amount = eval(parameter_expression_TBM, {}, parameter_registry)  # Evaluate the parameter expression to get the amount
                    print(f"Evaluated amount for '{ei_activity_name_TBM}': {amount}")
                    exchanges_dict[ei_activity_name_TBM] = dict(amount=amount)  # Structure the exchange as a dictionary with amount
                except Exception as e:
                    print(f"Error evaluating parameter expression '{parameter_expression_TBM}' for activity '{ei_activity_name_TBM}': {e}")
                    continue  # Skip this exchange if there's an error

            try:  # Now, update exchanges for the new activity
                print(exchanges_dict)  # Optional print for debugging
                new_activity.updateExchanges(exchanges_dict)
                print(f"Updated exchanges for activity: {modified_activity_name}")

            except Exception as e:
                print(f"Error updating exchanges for activity '{modified_activity_name}': {e}")
                continue
        except Exception as e:
            print(f"Error in copying or updating activity '{modified_activity_name}': {e}")
```

```python
for index, row in foregrounds_All[f"foreground_CGM_{USER_DB}"].iterrows():          # Iterate through the rows of the foreground sheet and create new activities
    
    ei_activity_name = row["EI activity name"]       # Read values from the Excel file
    location = row["loc"]                            # Location (optional)
    new_activity_name = row["LCA algebraic name"]    # the name of the new activity is defined in "LCA algebraic name"
    parameter_expression = row["parameter"]          # Name of the associated parameter
    modified_activity_name = row["modified"]         # name of the modified activity (if relevant)
    custom_process_name = row["Custom process name"] # name of the custom activity (if relevant)
   
    if pd.isna(ei_activity_name):  # Stop processing when 'EI activity name' is NaN
        print(f"Stopping processing at index {index} as 'EI activity name' is NaN.")
        break  # Exit the loop

    else:  
        ei_activity = external_functions.find_activity(ei_activity_name, location, 
                                    custom_process_name = custom_process_name, 
                                    modified_activity_name = modified_activity_name,
                                    custom_db=USER_DB) # Let's find the activity, either custom, modified or already present in ecoinvent
        
        if ei_activity is None:
            print(f"Skipping creation of {new_activity_name} due to unresolved activity issues.")
            continue  

        try:
            exchanges = {ei_activity: eval(parameter_expression, {}, parameter_registry)}  # Define exchanges
        except Exception as e:
            print(f"Error in parameter expression for activity '{new_activity_name}': {e}")
            continue

        try:
            agb.newActivity(USER_DB, new_activity_name, "unit", exchanges=exchanges)
            print(f"Activity created: {new_activity_name}")
        except Exception as e:
            print(f"Error creating activity '{new_activity_name}': {e}")
```

# Foreground generation

```python
if USER_DB == 'A0' or USER_DB == 'A1':
    
    CGM_and_Applicator_packaged_delivered = {} # Initialize an empty dictionary to store exchanges
    
    for index, row in foregrounds_All[f"foreground_CGM_{USER_DB}"].iterrows(): # Iterate through the main sheet to collect activities for the exchanges
        CGM_and_Applicator_packaged_delivered[agb.findActivity(row["LCA algebraic name"], db_name=USER_DB, loc="GLO")] = 1      
    agb.newActivity(USER_DB, "CGM and Applicator packaged and delivered", "unit", exchanges=CGM_and_Applicator_packaged_delivered) # Create the foreground

# ----------------------------------------------------------------------------------------------

elif USER_DB == 'A2':
    Applicator_packaged_delivered = {}
    Applicator_packaged_delivered_list = json.load(open('Foregrounds/Applicator_list_A2.txt'))
    for name in Applicator_packaged_delivered_list:
        Applicator_packaged_delivered[agb.findActivity(name, db_name=USER_DB, loc="GLO")] = 1
    Applicator_packaged_delivered_activity = agb.newActivity(USER_DB,"Applicator packaged delivered","unit", exchanges=Applicator_packaged_delivered)
    
    CGM_packaged_delivered = {}
    CGM_packaged_delivered_list = json.load(open('Foregrounds/CGM_list_A2.txt'))
    for name in CGM_packaged_delivered_list:
        CGM_packaged_delivered[agb.findActivity(name, db_name=USER_DB, loc="GLO")] = 1
    CGM_packaged_delivered_activity = agb.newActivity(USER_DB,"CGM packaged delivered","unit", exchanges=CGM_packaged_delivered)

# ----------------------------------------------------------------------------------------------

elif USER_DB == 'A3':
    Applicator_packaged_delivered = {}
    Applicator_packaged_delivered_list = json.load(open('Foregrounds/Applicator_list_A3.txt'))
    for name in Applicator_packaged_delivered_list:
        Applicator_packaged_delivered[agb.findActivity(name, db_name=USER_DB, loc="GLO")] = 1
    Applicator_packaged_delivered_activity = agb.newActivity(USER_DB,"Applicator packaged delivered","unit", exchanges=Applicator_packaged_delivered)

    CGM_packaged_delivered = {}
    CGM_packaged_delivered_list = json.load(open('Foregrounds/CGM_list_A3.txt'))
    for name in CGM_packaged_delivered_list:
        CGM_packaged_delivered[agb.findActivity(name, db_name=USER_DB, loc="GLO")] = 1
    CGM_packaged_delivered_activity = agb.newActivity(USER_DB,"CGM packaged delivered","unit", exchanges=CGM_packaged_delivered)

    Needles_packaged_delivered = {}
    Needles_packaged_delivered_list = json.load(open('Foregrounds/Needles_list_A3.txt'))
    for name in Needles_packaged_delivered_list:
        Needles_packaged_delivered[agb.findActivity(name, db_name=USER_DB, loc="GLO")] = 1
    Needles_packaged_delivered_activity = agb.newActivity(USER_DB,"Needles packaged delivered","unit", exchanges=Needles_packaged_delivered)    

# ----------------------------------------------------------------------------------------------

elif USER_DB == 'B':
    print('ok')
```

```python
# Used in Alternative A: N, the number of CGM used per year
nb_of_CGM = agb.newFloatParam('nb_of_CGM', default=31, min=26, max=37,  distrib=agb.DistributionType.TRIANGLE, description="None", label="nb of CGM")
parameter_registry['nb_of_CGM'] = nb_of_CGM

# Used in Alternative B: the number of blood test per year & the lifetime of the blood reader
nb_of_test = agb.newFloatParam('nb_of_test', default=730, min=365, max=2555, distrib=agb.DistributionType.TRIANGLE, description="None", label="nb of test")
parameter_registry['nb_of_test'] = nb_of_test

lt_accu_check = agb.newFloatParam( 'lt_accu_check', default=0.33, min=0.25, max=0.5, distrib=agb.DistributionType.TRIANGLE, description="None", label="lt_accu_check")
parameter_registry['lt_accu_check'] = lt_accu_check
```

```python
total_foreground_exchanges = {}

if USER_DB == 'A0' or USER_DB == 'A1':
    total_foreground_exchanges[agb.findActivity("CGM and Applicator packaged and delivered", db_name=USER_DB, loc="GLO")] = eval('nb_of_CGM', {}, parameter_registry)

if USER_DB == 'A2':
    total_foreground_exchanges[agb.findActivity("Applicator packaged delivered", db_name=USER_DB, loc="GLO")] = 1
    total_foreground_exchanges[agb.findActivity("CGM packaged delivered", db_name=USER_DB, loc="GLO")] = eval('nb_of_CGM', {}, parameter_registry)

if USER_DB == 'A3':
    total_foreground_exchanges[agb.findActivity("Applicator packaged delivered", db_name=USER_DB, loc="GLO")] = 1
    total_foreground_exchanges[agb.findActivity("CGM packaged delivered", db_name=USER_DB, loc="GLO")] = 2
    total_foreground_exchanges[agb.findActivity("Needles packaged delivered", db_name=USER_DB, loc="GLO")] = eval('nb_of_CGM', {}, parameter_registry)
```

# Reference Flow

```python
reference_flow = agb.newActivity(USER_DB, "reference flow", "unit", exchanges=total_foreground_exchanges)
```

# Impact Methods Choice

```python
# Retrieve impact categories (here from EF v3.1)
impacts_GWP = agb.findMethods(search="climate change", mainCat="EF v3.1") # GWP impact categories from EF 3.1
impacts_all = agb.findMethods(search="", mainCat="EF v3.1")               # All impact categories from EF 3.1
```

```python
absolute_impacts = agb.compute_impacts(reference_flow, impacts_all, functional_unit= 1)
absolute_impacts
```

# Contribution Analysis

```python
for index, row in foregrounds_All[f"foreground_CGM_{USER_DB}"].iterrows():
    activity_name = row["LCA algebraic name"]
    sub_assembly_label = row["Sub-assemblies 3"]  # choosing labels for activities -> life cycle phases : results per phase, 
    agb.findActivity(activity_name, db_name=USER_DB, loc="GLO").updateMeta(phase=sub_assembly_label, label=sub_assembly_label)
print("Finished labeling activities.")

df_impacts_axis = agb.compute_impacts(reference_flow, impacts_all, functional_unit=1, axis="label")
df_impacts_axis_ = df_impacts_axis.drop(index=['*sum*'])     # Drop the 'sum' 
df_sums = df_impacts_axis_.sum(axis=0)                       # Compute sums 
df_normalized = df_impacts_axis_.div(df_sums, axis=1) * 100  # Normalize data

# Create the plot
fig, ax = plt.subplots(figsize=(12, 10))
dark_palette = sns.color_palette("Spectral", n_colors=len(df_normalized.index))
df_normalized.T.plot(kind='bar', stacked=True, ax=ax, width=0.8, color=dark_palette)

# Customize plot
ax.set_title('Impact Categories with Contributions from Selected Labels (Normalized to 100%)')
ax.set_ylabel('Percentage (%)')
ax.set_xticklabels(df_normalized.columns, rotation=45, ha='right')
ax.legend(title="Labels", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
